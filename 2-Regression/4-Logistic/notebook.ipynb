## Pumpkin Varieties and Color

Load up required libraries and dataset. Convert the data to a dataframe containing a subset of the data: 

Let's look at the relationship between color and variety

```python
import pandas as pd
import numpy as np

full_pumpkins = pd.read_csv('../data/US-pumpkins.csv')

full_pumpkins.head()
```

```python
# Select columns to use
columns_to_select = ["City Name", "Package", "Variety", "Origin", "Item Size", "Color"]
pumpkins = full_pumpkins.loc[:, columns_to_select]
pumpkins.head()
```

```python
# Drop rows with missing values
pumpkins.dropna(inplace=True)
pumpkins.head()
```

Using a different plotting library: seaborn

```python
import seaborn as sns

# Specify colours for each value of the hue variable
palette = {
  "ORANGE": "orange",
  "WHITE": "wheat"
}

# Plot a bar plot to visualise how many pumpkins of each variety are orange or white
sns.catplot(
  data=pumpkins, y="Variety", hue="Color", kind="count", palette=palette
)
```

```python
# Look at different values of the "Item Size" column
pumpkins["Item Size"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Encode "Item Size" column using ordinal encoder
item_size_categories = [["sml", "med", "med-lge", "lge", "xlge", "jbo", "exjbo"]]
ordinal_features = ["Item Size"]
ordinal_encoder = OrdinalEncoder(categories=item_size_categories)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Encode all other features using one-hot encoding
categorical_features = ["City Name", "Package", "Variety", "Origin"]
categorical_encoder = OneHotEncoder(sparse_output=False)
```

```python
# Now we do data transformation
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers=[
  ("ord", ordinal_encoder, ordinal_features),
  ("cat", categorical_encoder, categorical_features)
])

# Get encoded features as pandas DataFrame
ct.set_output(transform="pandas")
encoded_features = ct.fit_transform(pumpkins)
encoded_features.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode "Color" column using label encoding
label_encoder = LabelEncoder()
encoded_label = label_encoder.fit_transform(pumpkins["Color"])
encoded_pumpkins = encoded_features.assign(Color=encoded_label)
encoded_pumpkins.head()
```

```python
# View the mapping between encoded and original values
list(label_encoder.inverse_transform([0, 1]))
```

```python
palette = {
  "ORANGE": "orange",
  "WHITE": "wheat"
}

# We need encoded "Item Size" column to use as the X axis values
pumpkins["Item Size"] = encoded_pumpkins["ord__Item Size"]

# Set up our catplot
g = sns.catplot(
  data=pumpkins,
  x="Item Size", y="Color", row="Variety",
  kind="box", orient="h",
  sharex=False, margin_titles=True,
  height=1.5, aspect=4, palette=palette
)

# Define X axis labels
g.set(xlabel="Item Size", ylabel="").set(xlim=(0, 6))
g.set_titles(row_template="{row_name}")
```

```python
# Use swarmplot to understand relationship between item size and colour
palette = {
  "0": "orange",
  "1": "wheat"
}

sns.swarmplot(x="Color", y="ord__Item Size", data=encoded_pumpkins, palette=palette)
```

Now we introduce logistic regression for the model

```python
from sklearn.model_selection import train_test_split

# X is encoded features
X = encoded_pumpkins[encoded_pumpkins.columns.difference(["Color"])]

#  y is encoded label
y = encoded_pumpkins["Color"]
```

```python
# Split to train/test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import f1_score, classification_report
from sklearn.linear_model import LogisticRegression

# Train logistic model on the pumpkin dataset
model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
# Evaluate model and print results
print(classification_report(y_test, predictions))
print("Predicted labels: ", predictions)
print("F1-score: ", f1_score(y_test, predictions))
```

To understand these results properly, need to understand/use confusion matrix

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions)
```

Says 162 accurate orange predictions (true negative), 22 accurate white predications (true positive). Also incorrectly predicted 4 as white (false positive), and 11 as orange (false negative)

### Confusion matrix

|        |   | Predicted |    |   |
|--------|---|-----------|----|---|
|        |   | 0         | 1  |   |
| Actual | 0 | TN        | FP |   |
|        | 1 | FN        | TP |   |

**Confusion Matrix table**


```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

y_scores = model.predict_proba(X_test)
# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_scores[:, 1])

# Plot ROC curve
fig = plt.figure(figsize=(6, 6))
# Plot diagonal 50% line
plt.plot([0, 1], [0, 1], "k--")
# Plot FPR and TPR acheived by our model
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

We can now calculate how good this ROC curve is with the score by calculating the area under the curve. Ideal is 1.

```python
auc = roc_auc_score(y_test, y_scores[:, 1])
print(auc)
```
